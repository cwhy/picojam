```python
import jax
import jax.numpy as jnp
from jax import random, grad, jit, value_and_grad, vmap
import optax
from typing import Tuple, Dict, Callable, Any, List, NamedTuple
from functools import partial
from datasets import load_dataset
from jax.nn import one_hot, relu, sigmoid
import matplotlib.pyplot as plt
import faiss
import numpy as np
import time

# Define data structures for type hints
class TrainState(NamedTuple):
    params: List[Dict[str, jnp.ndarray]]
    opt_state: optax.OptState
    step: int

class FaissState:
    def __init__(self, dataset: jnp.ndarray):
        # Precompute flattened dataset
        self.dataset_original = dataset
        self.dataset_flat_jax = dataset.reshape(dataset.shape[0], -1)
        self.dataset_flat_np = np.asarray(self.dataset_flat_jax, dtype=np.float32)
        
        # Build Faiss index
        self.index = faiss.IndexFlatL2(self.dataset_flat_np.shape[1])
        self.index.add(self.dataset_flat_np)

def load_mnist_data() -> Tuple[jnp.ndarray, jnp.ndarray, jnp.ndarray, jnp.ndarray]:
    """Load and preprocess MNIST dataset using Hugging Face datasets with JAX format."""
    print("Loading MNIST dataset from Hugging Face with JAX format...")
    
    # Load dataset from Hugging Face with JAX format
    mnist_dataset = load_dataset("mnist", cache_dir="~/.cache/huggingface/datasets").with_format("jax")
    
    # Extract train and test sets directly as JAX arrays
    train_data = mnist_dataset["train"]
    test_data = mnist_dataset["test"]
    
    # Extract images and reshape to 28x28
    x_train = jnp.array([img.reshape(28, 28) for img in train_data["image"]])
    y_train = jnp.array(train_data["label"])
    
    x_test = jnp.array([img.reshape(28, 28) for img in test_data["image"]])
    y_test = jnp.array(test_data["label"])
    
    # Normalize data
    x_train = x_train.astype(jnp.float32) / 255.0
    x_test = x_test.astype(jnp.float32) / 255.0
    
    # Convert labels to one-hot encoding
    y_train = one_hot(y_train, 10)
    y_test = one_hot(y_test, 10)
    
    print("x_train shape:", x_train.shape)
    print("y_train shape:", y_train.shape)
    print("x_test shape:", x_test.shape)
    print("y_test shape:", y_test.shape)
    
    return x_train, y_train, x_test, y_test

def init_mlp_params(rng: jnp.ndarray, layer_sizes: List[int]) -> List[Dict[str, jnp.ndarray]]:
    """Initialize parameters for a simple MLP."""
    keys = random.split(rng, len(layer_sizes) - 1)
    params = []
    
    for i, (in_size, out_size) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
        # Initialize weights with He initialization
        w_key, b_key = random.split(keys[i])
        w = random.normal(w_key, (in_size, out_size)) * jnp.sqrt(2.0 / in_size)
        b = jnp.zeros(out_size)
        params.append({"w": w, "b": b})
    
    return params

@jit
def mlp_forward(params: List[Dict[str, jnp.ndarray]], inputs: jnp.ndarray) -> jnp.ndarray:
    """Forward pass through the MLP."""
    activations = inputs
    
    # Apply all but the last layer with ReLU
    for layer_params in params[:-1]:
        activations = relu(jnp.dot(activations, layer_params["w"]) + layer_params["b"])
    
    # Apply the last layer with sigmoid for pixel values in [0, 1]
    final_output = sigmoid(jnp.dot(activations, params[-1]["w"]) + params[-1]["b"])
    
    return final_output

# Vectorize the forward pass for efficient batch processing
vmap_mlp_forward = jit(vmap(lambda p, x: mlp_forward(p, x), in_axes=(None, 0)))

@jit
def mse_loss(generated: jnp.ndarray, target: jnp.ndarray) -> jnp.ndarray:
    """Compute MSE loss between generated and target images."""
    return jnp.mean((generated - target) ** 2)

@jit
def compute_jax_loss(params: List[Dict[str, jnp.ndarray]], noise: jnp.ndarray, 
                     targets: jnp.ndarray) -> jnp.ndarray:
    """Compute loss for a batch using pure JAX operations."""
    # Generate images
    generated = mlp_forward(params, noise)
    
    # Reshape if needed
    if generated.shape[-1] == 784 and targets.shape[-1] != 784:
        generated = generated.reshape(generated.shape[0], 28, 28)
    
    # Compute MSE loss
    return mse_loss(generated, targets)

@partial(jit, static_argnums=(3,))
def update_step(state: TrainState, noise: jnp.ndarray, targets: jnp.ndarray, 
                optimizer: optax.GradientTransformation) -> Tuple[TrainState, Dict]:
    """JIT-compiled update step using pure JAX operations."""
    # Define loss function
    def loss_fn(params):
        return compute_jax_loss(params, noise, targets)
    
    # Compute gradients
    loss, grads = value_and_grad(loss_fn)(state.params)
    
    # Update parameters
    updates, opt_state = optimizer.update(grads, state.opt_state)
    params = optax.apply_updates(state.params, updates)
    
    # Create new state
    new_state = TrainState(
        params=params,
        opt_state=opt_state,
        step=state.step + 1
    )
    
    return new_state, {"loss": loss}

def hybrid_train_step(state: TrainState, noise: jnp.ndarray, faiss_state: FaissState, 
                      optimizer: optax.GradientTransformation) -> Tuple[TrainState, Dict]:
    """
    Hybrid training step that uses Faiss for nearest neighbor search and JAX for optimization.
    This ensures each generated image is matched to a unique image from the dataset.
    """
    # Generate images using JIT-compiled forward pass
    generated_images = mlp_forward(state.params, noise)
    
    # Reshape to flat 784-dimensional vectors for Faiss
    batch_size = noise.shape[0]
    generated_flat = generated_images.reshape(batch_size, 784)
    
    # Convert to numpy for Faiss
    generated_np = np.asarray(generated_flat, dtype=np.float32)
    
    # Find K nearest neighbors for each image
    # We retrieve K=batch_size+10 to have enough candidates to pick unique matches
    k = min(batch_size + 10, 50, faiss_state.dataset_original.shape[0])  # Limit K to reasonable number
    distances, indices = faiss_state.index.search(generated_np, k)
    
    # Initialize arrays to store the matched indices and distances
    matched_indices = np.zeros(batch_size, dtype=np.int32)
    matched_distances = np.zeros(batch_size, dtype=np.float32)
    
    # Track used indices to ensure uniqueness
    used_indices = set()
    
    # Assign unique nearest neighbors for each generated image (greedy algorithm)
    for i in range(batch_size):
        # Find the closest unused image for this generated image
        for j in range(k):
            candidate_idx = indices[i, j]
            if candidate_idx not in used_indices:
                matched_indices[i] = candidate_idx
                matched_distances[i] = distances[i, j]
                used_indices.add(candidate_idx)
                break
        else:
            # If all candidates are used, choose a random unused index
            # This is a fallback that should rarely happen if k is large enough
            available_indices = list(set(range(faiss_state.dataset_original.shape[0])) - used_indices)
            if available_indices:
                random_idx = np.random.choice(available_indices)
                matched_indices[i] = random_idx
                # Calculate distance for this fallback case
                img_diff = generated_flat[i] - faiss_state.dataset_flat_jax[random_idx]
                matched_distances[i] = float(jnp.sqrt(jnp.sum(img_diff**2)))
                used_indices.add(random_idx)
            else:
                # Extremely unlikely case: all dataset images are used
                # Just use the first match and accept the non-uniqueness
                matched_indices[i] = indices[i, 0]
                matched_distances[i] = distances[i, 0]
    
    # Get target images from dataset
    target_images = jnp.array([faiss_state.dataset_original[idx] for idx in matched_indices])
    
    # Update parameters using JIT-compiled update step
    new_state, metrics = update_step(state, noise, target_images, optimizer)
    
    # Add distance info to metrics
    metrics["avg_distance"] = float(np.mean(matched_distances))
    metrics["unique_targets"] = len(set(matched_indices))
    
    return new_state, metrics

def visualize_generated_images(params: List[Dict[str, jnp.ndarray]], faiss_state: FaissState, 
                              key: jnp.ndarray, n: int = 5):
    """
    Visualize generated images alongside their closest matches from the dataset.
    Ensures each generated image is matched to a unique dataset image.
    """
    # Generate random noise input
    noise_batch = random.normal(key, shape=(n, 100))
    
    # Generate images
    generated_images = mlp_forward(params, noise_batch)
    generated_images_reshaped = generated_images.reshape(n, 28, 28)
    
    # Prepare for Faiss (convert to flat vectors)
    generated_flat = generated_images.reshape(n, 784)
    generated_np = np.asarray(generated_flat, dtype=np.float32)
    
    # Find k nearest neighbors for each image
    k = min(n + 5, 20, faiss_state.dataset_original.shape[0])
    distances, indices = faiss_state.index.search(generated_np, k)
    
    # Initialize arrays to store the matched indices
    matched_indices = np.zeros(n, dtype=np.int32)
    
    # Track used indices to ensure uniqueness
    used_indices = set()
    
    # Assign unique nearest neighbors for each generated image
    for i in range(n):
        for j in range(k):
            candidate_idx = indices[i, j]
            if candidate_idx not in used_indices:
                matched_indices[i] = candidate_idx
                used_indices.add(candidate_idx)
                break
        else:
            # If all candidates are used, choose first one (happens rarely in visualization)
            matched_indices[i] = indices[i, 0]
    
    # Plot
    fig, axs = plt.subplots(n, 2, figsize=(6, 3*n))
    for i in range(n):
        if n == 1:
            curr_ax = axs
            ax1, ax2 = curr_ax[0], curr_ax[1]
        else:
            ax1, ax2 = axs[i, 0], axs[i, 1]
            
        ax1.imshow(generated_images_reshaped[i], cmap='gray')
        ax1.set_title(f"Generated {i+1}")
        ax1.axis('off')
        
        ax2.imshow(faiss_state.dataset_original[matched_indices[i]], cmap='gray')
        ax2.set_title(f"Closest Match (#{matched_indices[i]})")
        ax2.axis('off')
    
    plt.tight_layout()
    plt.show()

def main():
    print("Starting MNIST Generator with Unique Nearest Neighbor Loss")
    start_time = time.time()
    
    # Define master key for reproducibility
    master_key = random.PRNGKey(42)
    
    # Load MNIST data
    x_train, y_train, x_test, y_test = load_mnist_data()
    print(f"Data loaded in {time.time() - start_time:.2f} seconds")
    
    # Initialize Faiss state
    faiss_time = time.time()
    faiss_state = FaissState(x_train)
    print(f"Faiss index built in {time.time() - faiss_time:.2f} seconds")
    
    # Define MLP architecture
    layer_sizes = [100, 256, 512, 784]  # 100-dim noise -> 256 -> 512 -> 784 (28*28)
    
    # Initialize parameters
    init_time = time.time()
    key, subkey = random.split(master_key)
    params = init_mlp_params(subkey, layer_sizes)
    
    # Create optimizer
    learning_rate = 0.001
    optimizer = optax.adam(learning_rate)
    opt_state = optimizer.init(params)
    
    # Initialize training state
    state = TrainState(
        params=params,
        opt_state=opt_state,
        step=0
    )
    print(f"Model initialized in {time.time() - init_time:.2f} seconds")
    
    # Warm up the JIT compiler
    print("Warming up JIT compiler...")
    warmup_time = time.time()
    key, subkey = random.split(key)
    noise = random.normal(subkey, shape=(16, 100))
    dummy_targets = jnp.zeros((16, 28, 28))
    _ = update_step(state, noise, dummy_targets, optimizer)
    print(f"JIT warmup completed in {time.time() - warmup_time:.2f} seconds")
    
    # Training loop
    batch_size = 64
    num_epochs = 10
    steps_per_epoch = min(len(x_train) // batch_size, 100)  # Limit steps per epoch for demo
    
    print(f"Starting training with {num_epochs} epochs, {steps_per_epoch} steps per epoch")
    print(f"Using unique nearest neighbor matching for each batch of {batch_size} images")
    training_start = time.time()
    
    for epoch in range(num_epochs):
        epoch_start = time.time()
        epoch_loss = 0.0
        unique_targets_count = 0
        
        for step in range(steps_per_epoch):
            step_start = time.time()
            
            # Generate random noise
            key, subkey = random.split(key)
            noise = random.normal(subkey, shape=(batch_size, 100))
            
            # Update model
            state, metrics = hybrid_train_step(state, noise, faiss_state, optimizer)
            
            epoch_loss += metrics["loss"]
            unique_targets_count += metrics.get("unique_targets", 0)
            
            if step % 10 == 0:
                print(f"Epoch {epoch+1}/{num_epochs}, Step {step+1}/{steps_per_epoch}, " +
                      f"Loss: {metrics['loss']:.4f}, Avg Distance: {metrics.get('avg_distance', 0):.4f}, " +
                      f"Unique Targets: {metrics.get('unique_targets', 0)}/{batch_size}, " +
                      f"Step Time: {time.time() - step_start:.3f}s")
        
        avg_epoch_loss = epoch_loss / steps_per_epoch
        avg_unique_targets = unique_targets_count / steps_per_epoch
        print(f"Epoch {epoch+1}/{num_epochs} completed in {time.time() - epoch_start:.2f}s, " +
              f"Avg Loss: {avg_epoch_loss:.4f}, Avg Unique Targets: {avg_unique_targets:.1f}/{batch_size}")
        
        # Visualize some generated images
        if (epoch + 1) % 2 == 0 or epoch == 0:
            key, subkey = random.split(key)
            visualize_generated_images(state.params, faiss_state, subkey)
    
    print(f"Training completed in {time.time() - training_start:.2f} seconds")

if __name__ == "__main__":
    main()
```
